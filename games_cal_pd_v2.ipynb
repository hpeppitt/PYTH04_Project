```python
# Import libraries for data
import pandas as pd
import numpy as np
```

```python
# Specify the urls and tables to scrape
switch1 = pd.read_html("https://en.wikipedia.org/wiki/List_of_Nintendo_Switch_games", header=[1])[0]
switch2 = pd.read_html("https://en.wikipedia.org/wiki/List_of_Nintendo_Switch_games_(M-Z)", header=[1])[0]
xbox1 = pd.read_html("https://en.wikipedia.org/wiki/List_of_Xbox_One_games_(A-L)", header=[1])[1]
xbox2 = pd.read_html("https://en.wikipedia.org/wiki/List_of_Xbox_One_games_(M-Z)", header=[1])[1]
ps41 = pd.read_html("https://en.wikipedia.org/wiki/List_of_PlayStation_4_games", header=[1])[0]
ps42 = pd.read_html("https://en.wikipedia.org/wiki/List_of_PlayStation_4_games_(M-Z)", header=[1])[0]
# PC A - "https://en.wikipedia.org/wiki/List_of_PC_games_(A)"
# PC Z - "https://en.wikipedia.org/wiki/List_of_PC_games_(Z)"
# PC 0-9 - "https://en.wikipedia.org/wiki/List_of_PC_games_(Numerical)"
```

```python
# Check DataFrame results
#switch1
#switch2
#xbox1
#xbox2
#ps41
#ps42
```

```python
# Append column for platform
switch1['Platform'] = 'Nintendo Switch'
switch2['Platform'] = 'Nintendo Switch'
xbox1['Platform'] = 'Xbox One'
xbox2['Platform'] = 'Xbox One'
ps41['Platform'] = 'Playstation 4'
ps42['Platform'] = 'Playstation 4'
```

```python
# Delete unnessacary columns
switch1.drop(['Ref.'], axis=1)
switch2.drop(['Ref.'], axis=1)
xbox1.drop(['Ref.','Addons'], axis=1)
xbox2.drop(['Ref.','Addons'], axis=1)
ps41.drop(['Ref.','Addons'], axis=1)
ps42.drop(['Ref.','Addons'], axis=1)
```

```python
# Create new columns to extract date values
```

```python
# Concatenate all tables into one
df = pd.concat([switch1, switch2, xbox1, xbox2, ps41, ps42], ignore_index=True, sort=False)
df.to_csv ('df_v1.csv', index=False, header=True)
```

```python
# Create new alt columns to seperate non-date values
# Values contain [case sensitive]: Unreleased, Asia, China, TBA
temp=df.JP.fillna('NaN')
df['JP_alt'] = np.where(temp.str.contains('NaN'), '',
                    pd.np.where(temp.str.contains('Unreleased'), 'Unreleased',
                    pd.np.where(temp.str.contains('Asia'), 'Asia Exclusive',
                    pd.np.where(temp.str.contains('China'), 'China Exclusive',
                    pd.np.where(temp.str.contains('TBA'), 'TBA', '')))))

temp=df.EU.fillna('NaN')
df['EU_alt'] = np.where(temp.str.contains('NaN'), '',
                    pd.np.where(temp.str.contains('Unreleased'), 'Unreleased',
                    pd.np.where(temp.str.contains('Asia'), 'Asia Exclusive',
                    pd.np.where(temp.str.contains('China'), 'China Exclusive',
                    pd.np.where(temp.str.contains('TBA'), 'TBA', '')))))

temp=df.NA.fillna('NaN')
df['NA_alt'] = np.where(temp.str.contains('NaN'), '',
                    pd.np.where(temp.str.contains('Unreleased'), 'Unreleased',
                    pd.np.where(temp.str.contains('Asia'), 'Asia Exclusive',
                    pd.np.where(temp.str.contains('China'), 'China Exclusive',
                    pd.np.where(temp.str.contains('TBA'), 'TBA', '')))))

temp=df.PAL.fillna('NaN')
df['PAL_alt'] = np.where(temp.str.contains('NaN'), '',
                    pd.np.where(temp.str.contains('Unreleased'), 'Unreleased',
                    pd.np.where(temp.str.contains('Asia'), 'Asia Exclusive',
                    pd.np.where(temp.str.contains('China'), 'China Exclusive',
                    pd.np.where(temp.str.contains('TBA'), 'TBA', '')))))

# Create new date columns to seperate date values
# Values that don't contain [case sensitive]: Unreleased, Asia, China, TBA

df[['Title', 'JP', 'JP_alt', 'NA', 'NA_alt', 'PAL', 'PAL_alt', 'EU', 'EU_alt']]
```
