```python
###import libraries
import pandas as pd
import numpy as np
from datetime import datetime
```

```python
### Specify the urls and tables to scrape
switch1 = pd.read_html("https://en.wikipedia.org/wiki/List_of_Nintendo_Switch_games", header=[1])[0]
switch2 = pd.read_html("https://en.wikipedia.org/wiki/List_of_Nintendo_Switch_games_(M-Z)", header=[1])[0]
xbox1 = pd.read_html("https://en.wikipedia.org/wiki/List_of_Xbox_One_games_(A-L)", header=[1])[1]
xbox2 = pd.read_html("https://en.wikipedia.org/wiki/List_of_Xbox_One_games_(M-Z)", header=[1])[1]
ps41 = pd.read_html("https://en.wikipedia.org/wiki/List_of_PlayStation_4_games", header=[1])[0]
ps42 = pd.read_html("https://en.wikipedia.org/wiki/List_of_PlayStation_4_games_(M-Z)", header=[1])[0]
# PC A - "https://en.wikipedia.org/wiki/List_of_PC_games_(A)"
# PC Z - "https://en.wikipedia.org/wiki/List_of_PC_games_(Z)"
# PC 0-9 - "https://en.wikipedia.org/wiki/List_of_PC_games_(Numerical)"
```

```python
### Create refrence lists/dictionaries for dataframes
platforms = [switch1, 
             switch2, 
             xbox1, 
             xbox2,
             ps41, 
             ps42
            ]

non_dates = ['Unreleased',
             'Asia',
             'China',
             'TBA'
             ]

locales = ['JP' , 'NA', 'PAL']
```

```python
class DFFormat():
    def __init__(self, non_dates, locales, platform):
        self.non_dates = non_dates
        self.locales = locales
        self.platform = platform
        
    def non_date_columns(self, non_dates, locales, platform):
        for non_date in self.non_dates:
            for locale in self.locales:
                for row in platform[locale]:
                    temp = platform[locale].fillna('')
                    platform[f"{locale}_{non_date}"] = np.where(temp.str.contains(non_date), non_date,'')
```

```python
def platform_columns(platforms):
    for platform in platforms:
        print(platform)
        platform_formatted = DFFormat(non_dates, locales, platform)
        new_platforms = platform_formatted.non_date_columns(non_dates, locales, platform)
        return new_platforms
```

```python
def IsNumber(s):
    try: 
        int(s)
        return True
    except ValueError:
        return False

def ConvertToDatetime(date):
    date=date.split(" ") #split by space
    if(IsNumber(date[0])): #is of the form dd month year
        if(len(date[1])==3): #if month is for form Jan,Feb...
            datetime_object = datetime.strptime(" ".join(date), '%d %b %Y')
        else: #if month is for form January ,February ...
            datetime_object = datetime.strptime(" ".join(date), '%d %B %Y')
    else:  #is of the form month date year
        if(len(date[0])==3): #if month is for form Jan,Feb...
            datetime_object = datetime.strptime(" ".join(date), '%b %d %Y')
        else: #if month is for form January ,February ...
            datetime_object = datetime.strptime(" ".join(date), '%B %d %Y')
    return datetime_object
```
